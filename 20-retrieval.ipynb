# 20 - Retrieval: test semantic search & prepare context for LLM


```python
# imports and reload env
from dotenv import load_dotenv
import os
load_dotenv()
from sentence_transformers import SentenceTransformer
from pinecone import Pinecone
embed_model = SentenceTransformer("all-MiniLM-L6-v2")
pc = Pinecone(api_key=os.getenv("PINECONE_API_KEY"))
index = pc.Index(os.getenv("PINECONE_INDEX"))
print("Index ready:", os.getenv("PINECONE_INDEX"))
```

```python
# Query function that returns concatenated context
def get_context(query, k=5):
    qvec = embed_model.encode(query).tolist()
    res = index.query(vector=qvec, top_k=k, include_metadata=True)
    texts = [m["metadata"]["text"] for m in res.get("matches",[])]
    return "\n\n".join(texts)

# Test
qry = "severe headache and vomiting"
ctx = get_context(qry, k=3)
print("Context length:", len(ctx))
print(ctx[:800])
```
